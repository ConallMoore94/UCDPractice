```python
#creates a function that recieves a specific column and returns as a new dataframe
def split_multicolumn(genres):
    outcome_df = genres.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in genres[genres.notnull()].iteritems():
        # Break each value into list of options
        for option in value.split(','):
            # Add the option as a column to outcome
            if not option in outcome_df.columns:
                options.append(option)
                outcome_df[option] = False
            # Mark the value in the option column as True
            outcome_df.at[idx, option] = True
    return outcome_df[options]
```

```python
#create a new multicolumn with the splitmulticolumn function
genres_df = split_multicolumn(streaming_data.Genres)

#performing a sum operation of all True values
streaming_genres = genres_df.sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,8))
plt.title('Genres via Streaming Services')
sns.barplot(x=streaming_genres.values, y=streaming_genres.index)

```
